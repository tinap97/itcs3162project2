Pre-Processing/Cleaning

```python
df.head()
```

```python
for col in df.columns:
    print(f"{col}:\n {df[col].value_counts()}\n")
```

```python
# dropping features: BMI, Income, Education, Age, Sex, General Health,MentHlth, PhysHlth, DiffWalk
df.drop('BMI', axis = 1, inplace=True)
df.drop('GenHlth', axis = 1, inplace=True)
df.drop('Age', axis = 1, inplace=True)
df.drop('Education', axis = 1, inplace=True)
df.drop('Income', axis = 1, inplace=True)
#df.drop('MentHlth', axis = 1, inplace=True)
#df.drop('PhysHlth', axis = 1, inplace=True)
#df.drop('DiffWalk', axis = 1, inplace=True)
```

```python
sns.countplot(x=df["Diabetes_binary"])
```

```python
#Set our X and y values
X = df.drop(columns = ['Diabetes_binary'])
y = df['Diabetes_binary']
```

```python
X.shape, y.shape
```

```python
X
```

```python
y
```

```python
#separate our data into a training and testing set
from sklearn.model_selection import train_test_split
```

```python
#Using only 20% of the data as test size
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.2 )
```

```python
print("X_train", X_train.shape)
print("X_test", X_test.shape)
print("y_train", y_train.shape)
print("y_test", y_test.shape)
```

```python
#Modeling
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier() 
```

```python
clf = clf.fit(X_train, y_train)
```

```python
#model is now trained
#Evaluate
```

```python
predicted = clf.predict(X_test)
```

```python
predicted
```

```python
#get accuracy score
clf.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt
```

```python
#create vizualization for decision tree
fig, axes = plt.subplots(nrows = 1,ncols = 1, figsize = (7,5), dpi=900)
tree.plot_tree(clf, max_depth = 3, feature_names = X.columns, filled=True)
plt.show()
```

```python
#feature importance array
```

```python
f = clf.feature_importances_ 
f = pd.Series(data = f, index = X.columns) 
f.sort_values(ascending=False, inplace=True) 
```

```python
#create bar plot
plt.figure(figsize=(12, 6))
chart = sns.barplot(x=f, y=f.index, palette=sns.color_palette("pastel", n_colors=len(f)))
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```
